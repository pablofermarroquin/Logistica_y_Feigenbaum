```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
t= np.linspace(0, 0.05, 300, endpoint = True)

k=1000
R=0.2
C=0.9
No=100
C=No/(k-No)
N=(C*k*np.exp(t*k*R))/(1+C*k*np.exp(t*k*R))

plt.xlabel("Tiempo (s)")
plt.ylabel("Posición (m)")
plt.plot(t,N,c="g")
plt.title("\nPoblación vrs. tiempo") 
```

```python
N=[]
k=1000
R=0.2
C=0.9
No=1500
C=No/(k-No)
N.append(100)
t=[]
t.append(0)
i=0
while i<0.05:
  pos=len(N)-1
  N_pre=N[pos]
  N.append(N_pre+ R*N_pre*(k-N_pre)*0.001)
  t.append(i)
  i+=0.001

plt.xlabel("Tiempo (s)")
plt.ylabel("Posición (m)")
plt.plot(t,N,c="g")
plt.title("\nPoblación vrs. tiempo") 
plt.show()
#N
```

```python
Rs=np.arange(2.7, 4, 0.005).tolist()

yn = 0.5
y_mas1 = []

for R in Rs:
    y_temp = []
    for i in range(200):
        yn = R * yn * (1.0 - yn)
        y_temp.append(yn)
    y_mas1.append(y_temp)

plt.plot(Rs,y_mas1, '.k')

plt.show()
```

```python
def calc(lim, d, inicio):
    N = 2 ** d
    for _ in range(lim):
        x = 0
        Xp = 0
        for _ in range(N):
            Xp = 1 - 2 * Xp * x
            x = inicio - x ** 2
        inicio -= x / Xp
    return inicio


l1 = [0, 1]
cte = 3.2
for i in range(2, 14 + 1):
    aprox = l1[-1] + ((l1[-1] - l1[-2]) / cte)
    l1.append(calc(500, i, aprox))
    Fei = (l1[-2] - l1[-3]) / (l1[-1] - l1[-2])
print("Constante de Feigenbaum: "+str(Fei))
```
